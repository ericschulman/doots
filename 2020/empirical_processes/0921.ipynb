```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = np.linspace(1,10,10)
print(c)
```

```python
def prokhorov(c,n,sigma):
    t = np.linspace(0,1,20)
    return np.exp(-3*t/(4*c) *np.log(1+ 2*t*c/(3*n*sigma**2)  ))

c = np.linspace(1,5,10)
n = 5
sigma = .1

for ci in c:
    plt.plot(np.linspace(0.,1.,20),prokhorov(ci,n,sigma),label=ci)

plt.legend()
```
